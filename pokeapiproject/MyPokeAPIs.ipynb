```python
import requests as r
lugia = r.get("https://pokeapi.co/api/v2/pokemon/lugia")
print(lugia.status_code)
```

```python
if lugia.status_code == 200:
    pokedex = lugia.json() #converts 
    #print(pokedex)
    lugiapokedex = {}
    
    lugiapokedex["name"] = pokedex['name'] #value
    
    lugiaattacks = [] # write a list if it has more than one ability
    
    lugiatypes = []
    #abilities is a key which gives us access to a list that holds multiple dictionaries
    for ability in pokedex['abilities']:
        lugiaattacks.append(ability['ability']['name'])#ea. ability has 3 key value pairs
        lugiapokedex['abilities'] = lugiaattacks 
        
    lugiapokedex['weight'] = pokedex['weight']
    
    for types in pokedex['types']:
        lugiatypes.append(types['type']['name'])
        lugiapokedex['types'] = lugiatypes
    
    print(lugiapokedex)
```

```python
def pokefunct(pokemon):
    import requests as r
    req = r.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon}')
    if req.status_code == 200:
        api = req.json()
    else:
        return req.status_code # send back the error code if something went wrong
    
    pokedict = {
        'name': api['name'],
        'types': [x['type']['name'] for x in api['types']],
        'abilities': [x['ability']['name'] for x in api['abilities']],
        'weight': api['weight']
    }
    
    return pokedict

lugia = pokefunct('lugia')
print(lugia)
zap = pokefunct('zapdos')
print(zap)
```

```python
pokemon = ['lugia', 'zapdos', 'articuno', 'entei','mew','mewtwo','moltres','raikou','suicune','ho-oh','regirock','regice','registeel','latias','latios','kyogre','groudon','rayquaza','pikachu','spheal']

monlist = []
for name in pokemon:
    monlist.append(pokefunct(name))
    
monlist = [pokefunct(name) for name in pokemon]

# how are we printing with no print?
# it is a thing that jupyter notebook does -> it "prints" the output of the final line of code in any code block
import pprint
pp = pprint.PrettyPrinter(indent=2)
pp.pprint(monlist)
# pprint is also a package for printing things in easier-to-read formats
```

```python
types = set()
for i in monlist:
    for t in i['types']:
        types.add(t) # .add() for a set is just like .append() for a list

pokemonbytype = {}
for t in types:
    pokemonbytype[t] = {pokemon['name']:pokemon for pokemon in monlist if t in pokemon['types']}


# include a pokemon in my water type section if it has water in it's typelist
pokemonbytype['water'] = {pokemon['name']:pokemon for pokemon in monlist if 'water' in pokemon['types']}

pokemonbytype
```
